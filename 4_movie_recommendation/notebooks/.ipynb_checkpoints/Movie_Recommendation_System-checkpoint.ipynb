# User Based Filtering 

## Build User Item Matrix 

```python
# Using pivot to get users as rows so we can compare them,we need to replace Nans with zero[ no rating] becasue cosine_similarity can't do math with NaN
user_item_matrix=ratings.pivot(index='user_id', columns='movie_id', values='rating').fillna(0)
user_item_matrix.head()
```

```python
# Compute User Similarity ( has values between zero[not similar] and one[identical])
similar_users= cosine_similarity(user_item_matrix)

#Convert the similarity matrix into a DataFrame 
similar_users_df= pd.DataFrame(
            similar_users, 
    index= user_item_matrix.index,
    columns= user_item_matrix.index
)

similar_users_df.head()
```

# Generate Recommendations 

```python
def get_similar_users(user_id, similar_users_df): 
    top_ten_similar= similar_users_df.loc[user_id].sort_values(ascending= False)[1:11]
    return top_ten_similar
```

```python
def recommend_movies(user_id, num_recommendations, users_df= similar_users_df, ratings_df= ratings, movies_df= movies):

    """ 
    Generate recommendations for a user

    Returns: Movie titles the user should watch
    """
    top_ten_similar= get_similar_users(user_id, users_df)

    # Filter ratings from top ten similar users 
    filtered_ratings= ratings_df[ratings_df['user_id'].isin(top_ten_similar.index)].copy() 

    #Add Similarity Scores 
    filtered_ratings['similarity']= filtered_ratings['user_id'].map(top_ten_similar)
    #Calculate Weighted Ratings [We multiply rating by similarity because we want to trust the opinions of people similar to us ]
    filtered_ratings['weighted_ratings']= filtered_ratings['rating'] * filtered_ratings['similarity']

    #Aggregate by Movie
    movie_groups= filtered_ratings.groupby('movie_id').agg(
        {
            'weighted_ratings':'sum',
            'similarity':'sum'
        }
    )
    
    #We divide the sum of weighted ratings by the sum of similarities to get a rating that accounts for how similar each person is to us
    movie_groups['predicted_ratings']= movie_groups['weighted_ratings'] / movie_groups['similarity']

    #Get User's Already-Watched Movies [finds all the movies that the user has already watched/rated, so we don't recommend them again]
    users_watched_movies= ratings_df[ratings_df['user_id']== user_id]['movie_id']

    #Remove Already-Watched Movies [recommed_list: This is filtered DataFrame with unwatched movies, where movie_id is index and columns are weighted_ratings, similarity, predicted_ratings]
    recommed_list= movie_groups[~ movie_groups.index.isin(users_watched_movies)]

    #Get Top N Recommendations [ will give me as index the movie id and predicted ratings as values]
    top_n= recommed_list['predicted_ratings'].sort_values(ascending= False).head(num_recommendations)

    #Get Movie Title [This will give me movie id as index and titles as values]
    movie_titles= movies_df.set_index('movie_id').loc[top_n.index, 'title']

    return movie_titles.values

   
   
```

```python
 #Testing the function 

print("User 1:")
print(recommend_movies(1, 5))

print("\nUser 10:")
print(recommend_movies(42, 4))

print("\nUser 100:")
print(recommend_movies(900, 3))
```

**Let's Test how Good these Recommendations**

```python
def split_data_per_user(ratings_df= ratings, test_size=0.2, min_ratings=5):
    """ So by splitting per user, we guarantee: Every user in the test set also has data in the train set """


    #Inner Function runs once per user group
    def test_split(group_user_id):
        if len(group_user_id) < min_ratings:
            return pd.DataFrame() #Because groupby().apply() expects a DataFrame/Series so returning empty DF ensures consistent output shape
            
        #We ignore the train part because well reconstruct train later from the full set
        _, test = train_test_split(group_user_id, test_size=test_size, random_state=42)
        return test  # This returns only the test portion for this user


    
    test_set = ratings_df.groupby('user_id').apply(test_split).reset_index(drop=True)  # Using reset_index to Clean up the multi-index created by groupby().apply()
    train_set = ratings_df[~ratings_df.index.isin(test_set.index)]
    return train_set, test_set
```

```python
# Do this ONCE, before evaluating any user
train_ratings, test_ratings = split_data_per_user(ratings, test_size=0.2, min_ratings=5)

print(train_ratings.shape)
print(test_ratings.shape)
```

```python

train_ratings['rating_normalized'] = train_ratings.groupby('user_id')['rating'].transform(lambda x: x - x.mean())

# rating_normalized instead of 'rating'
user_item_matrix  = train_ratings.pivot(index='user_id', columns='movie_id', values='rating_normalized').fillna(0)
```

```python
# Build user-item matrix from TRAIN data only
# train_data_only_matrix = train_ratings.pivot(index='user_id', columns='movie_id', values='rating').fillna(0)

#Compute Similarity 
train_user_similarity = cosine_similarity(user_item_matrix)
train_user_similarity_df = pd.DataFrame(
    train_user_similarity,
    index=user_item_matrix.index,
    columns=user_item_matrix.index
)
```

```python
def evaluate_recommendations(user_id, train_df= train_ratings, test_df= test_ratings, similar_df= train_user_similarity_df, movies_df=  movies, n_recommed=5): 
    """
    TEST how good the recommendation system
    Returns: Metrics (precision score) showing accuracy
    """
    # Get this user's TRAIN and TEST ratings
    user_train= train_df[train_df['user_id']== user_id]
    user_test= test_df[test_df['user_id']== user_id]

    #Skip users with no test data or no similarity profile
    if user_test.empty: 
        return None 
    if user_id not in similar_df.index: 
        return None 
    
   

    #Generate Recommendations (Using Train Data)  
    
    top_similar = similar_df.loc[user_id].sort_values(ascending=False)[1:11]


    # Keep only users with similarity > 0.1 
    top_similar = top_similar[top_similar > 0.1]
    if len(top_similar) < 2:  # Require multiple similar users for reliable predictions
        return None



    #Get Ratings From Similar Users 
    similar_ratings= train_df[train_df['user_id'].isin(top_similar.index)].copy()

    
    similar_ratings['similarity']= similar_ratings['user_id'].map(top_similar) 
    similar_ratings['weighted_ratings']= similar_ratings['rating'] * similar_ratings['similarity']


    # grouped_movies= similar_ratings.groupby('movie_id').agg({'similarity': 'sum', 'weighted_ratings':'sum'})
    # grouped_movies['pred_score']= grouped_movies['weighted_ratings'] / grouped_movies['similarity']

    grouped_movies = similar_ratings.groupby('movie_id').agg({
    'similarity': 'sum',
    'weighted_ratings': 'sum',
    'user_id': 'count'  # counts how many similar users rated this movie
    })
    grouped_movies.rename(columns={'user_id': 'similar_count'}, inplace=True)
    
    # Require at least 2 similar users to have rated the movie
    grouped_movies = grouped_movies[grouped_movies['similar_count'] >= 2]
    
    
    grouped_movies['pred_score'] = grouped_movies['weighted_ratings'] / grouped_movies['similarity']


    movie_popularity = train_ratings['movie_id'].value_counts()
    grouped_movies['popularity'] = grouped_movies.index.map(movie_popularity).fillna(0)
    
    # By adding 0.001 * popularity, we slightly boost more popular movies only when scores are equal
    grouped_movies['final_score'] = grouped_movies['pred_score'] + 0.001 * grouped_movies['popularity']
    
    
    

    # Remove movies the user already rated in Train
    watched= set(user_train['movie_id'])

    # Filter and Get Top Recommendations 
    recommendations= grouped_movies[~grouped_movies.index.isin(watched)]

    if recommendations.empty:
        return None 
    # top_recommendations= recommendations['pred_score'].sort_values(ascending=False).head(n_recommed)
    top_recommendations =  recommendations['final_score'].sort_values(ascending=False).head(n_recommed)





    # Check How Good the Recommendations Are( Evaluation)
    liked_test_movies= set(user_test[user_test['rating']>= 4]['movie_id'])
    
   # Compute Percision --> Of all the things I suggested, how many were good
    recommend_movies= set(top_recommendations.index)
    hits= recommend_movies & liked_test_movies
    percision= len(hits) / len(top_recommendations)
   
    
    return {
        'user_id': user_id,
        'precision': percision,
        'recommended': list(top_recommendations.index),
        'liked_by_user': list(liked_test_movies),
        'hits_count': len(hits),
        'recommendations_count': len(top_recommendations)
    }
    
    
```

```python
# Evaluate first 20 test users

results = []
skipped = 0

for uid in test_ratings['user_id'].unique()[:20]:
    res = evaluate_recommendations(uid)
    if res:
        results.append(res)
    else:
        skipped += 1

avg_prec = sum(r['precision'] for r in results) / len(results) if results else 0
print(f"Average Precision@5 over {len(results)} users: {avg_prec:.3f}")
print(f"Skipped {skipped} users")
```

## Item -Item Matrix

```python
# Building item-item matrix 
item_user_matrix = user_item_matrix.T 

# Compute cosine similarity between movies to answer the question do users who liked movie A also tend to like movie B
item_similarity= cosine_similarity(item_user_matrix)


# Convert to DataFrame
item_similarity_df= pd.DataFrame (
    item_similarity, 
    index= item_user_matrix.index, 
    columns=item_user_matrix.index 
)

item_similarity_df.head()
```

```python
def predict_movie_score(user_id, movie_id, train_df= train_ratings, similar_df= item_similarity_df):

    if movie_id not in similar_df.index: 
        return 0.0
    
     # Get user's ratings
    user_ratings= train_df[train_df['user_id']== user_id]
    if user_ratings.empty:
        return 0.0 
        
    rated_movies= user_ratings.set_index('movie_id')['rating_normalized']

    user_mean = user_ratings['rating'].mean()

   
    
    # Get similarities between target movie and all rated movies
    similar_scores= similar_df.loc[movie_id,rated_movies.index ]
    similar_scores= similar_scores[similar_scores > 0]

    if len(similar_scores)==0:
        # return 0.0
        return user_mean



    weighted_sum= (similar_scores * rated_movies.loc[similar_scores.index]).sum()
    similarity_sum= similar_scores.sum()

    pred_score_norm= weighted_sum / similarity_sum
    pred_score = pred_score_norm + user_mean

    return pred_score



```

```python
# Test prediction for User 1 on Movie 1
pred = predict_movie_score(1, 1)
print(f"Predicted score for User 1 on Movie 1: {pred:.3f}")
```

```python
def recommed_items_based(user_id, train_df=train_ratings, similar_df=item_similarity_df, movies_df= movies, n_recommed=5): 
    """
        Generate item-based recommendations for a user.
        Returns: list of (movie_id, predicted_score)
    """

    user_ratings = train_df[train_df['user_id'] == user_id]
    rated_movies_id_by_user= set(user_ratings['movie_id'])


    all_movies= similar_df.index.tolist() 


    suggested_movies= [ id for id in all_movies if id not in rated_movies_id_by_user ]

    if not suggested_movies:
        return []


    predictions=[] 
    
    for movie_id in suggested_movies:
        pred_score= predict_movie_score(user_id, movie_id)
        predictions.append((movie_id, pred_score))  

    predictions.sort(key= lambda x: x[1], reverse= True)
    
    movie_popularity = train_df['movie_id'].value_counts()
    
    predictions_with_pop = []
    for movie_id, pred_score in predictions:
        pop = movie_popularity.get(movie_id, 0)
        final_score = pred_score + 0.3 * pop  # boost popular movies
        predictions_with_pop.append((movie_id, final_score))
    
    # Sort by final_score
    predictions_with_pop.sort(key=lambda x: x[1], reverse=True)
    return predictions_with_pop[:n_recommed]

    # return predictions[:n_recommed]
```

```python
recs = recommed_items_based(1)
print("Top 5 item-based recommendations for User 1:")
for mid, score in recs:
    title = movies[movies['movie_id'] == mid]['title'].iloc[0] if mid in movies['movie_id'].values else "Unknown"
    print(f"Movie {mid}: {title} (score: {score:.2f})")
```

```python
def evaluate_item_based_recommendations(user_id, train_df, test_df, item_sim_df, movies_df, n_recommed=5):
    """
    Evaluate item-based recommendations for a user.
    """
    # Get user's TRAIN and TEST ratings
    user_train = train_df[train_df['user_id'] == user_id]
    user_test = test_df[test_df['user_id'] == user_id]

    if user_test.empty:
        return None

    # Generate recommendations
    recs = recommed_items_based(user_id, train_df, item_sim_df, movies_df, n_recommed)
    if not recs:
        return None

    recommended_movie_ids = [mid for mid, _ in recs]

    # Check how many were liked in test set
    liked_test_movies = set(user_test[user_test['rating'] >= 4]['movie_id'])
    hits = set(recommended_movie_ids) & liked_test_movies
    precision = len(hits) / len(recommended_movie_ids)

    return {
        'user_id': user_id,
        'precision': precision,
        'recommended': recommended_movie_ids,
        'liked_by_user': list(liked_test_movies),
        'hits_count': len(hits),
        'recommendations_count': len(recommended_movie_ids)
    }
```

```python
# Evaluate first 20 users
results = []
for uid in test_ratings['user_id'].unique()[:20]:
    res = evaluate_item_based_recommendations(uid, train_ratings, test_ratings, item_similarity_df, movies)
    if res:
        results.append(res)

avg_prec = sum(r['precision'] for r in results) / len(results) if results else 0
print(f"Average Precision@5 over {len(results)} users: {avg_prec:.3f}")
```
